# Heuristic Optimization

- **Algorithm** is a sequence of instructions which provides a procedure to solve a computational problem
- **A computational problem P** is a class of algorithmically-doable tasks
- Set of instances IP (Input domain set)
     - For each x in IP , there is an associated set of (feasible) solutions solP(x)
     - The algorithm solving P returns one of the solutions in solP(x)
     
_______________________________________________________________________________________________________________________________

- Types of computational problems: 
 1. Enumeration problem: find all solutions in solP(x)
 2. Counting problem: count how many solutiuons exists in solP(x)
 3. Decision problem: determine whether the set solP(x) is empty or not
 4. Optimization problem: Find a solution in solP(x) maximizing or minimizing a given function
 _______________________________________________________________________________________________________________________________
 
**Optimization Problems**

- Searching the best (based on an objective function) configuration of a set of variables among all feasible (based on constraints) configurations
- Types of optimization problems:
    - Real valued variables
    - Discrete valued variables
        - Combinatorial Optimization
 _______________________________________________________________________________________________________________________________

*Combinatorial Optimization*

- *Most of* the combinatorial optimization problems are *NP-Hard*
- A *problem Q* is *NP-hard* *if all problems in NP* can be *reduced to Q*
- *NP-Hard* problems are *at least as* *hard as NPComplete* problems
- *There are NP-Hard* problems that are *not NPComplete*
- *There are no known polynomial-time algorithms for* *NP-Hard* problems!

**Meta-heuristics**

“A metaheuristic is a set of concepts that can be used to define heuristic methods that **can be applied to a wide set of different problems**. In other words, a metaheuristic can be seen as a general algorithmic framework which **can be applied to different optimization problems with relatively few modifications to make them adapted to a specific problem.”** 

“A metaheuristic is an **iterative master process** that guides and modifies the operations of subordinate heuristics **to efficiently produce high-quality solutions**. It may **manipulate a complete (or incomplete) single solution or a collection of solutions at each iteration**. The subordinate heuristics may be high (or low) level procedures, or a simple local search, or just a construction method.”
